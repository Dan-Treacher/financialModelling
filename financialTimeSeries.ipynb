```python
key_dir = 'C:\\Users\\Dan\\Documents\\Python Scripts\\apiKeys\\'
apiKey = open(key_dir + 'alpaca_apiKey.txt', 'r').read()
secretKey = open(key_dir + 'alpaca_secretKey.txt', 'r').read()
endPoint = 'https://paper-api.alpaca.markets'  # Specify this for demo account

# Define companies of interest
tickers = ['AXP', 'AAPL', 'BA', 'CAT', 'CVX', 'CSCO', 'DIS', 'DOW', 'XOM',
           'HD', 'IBM', 'INTC']

# Connect to api and pull data for the defined tickers
api = tradeapi.REST(apiKey, secretKey, endPoint, api_version='v2')

# Define the dictionary that will hold all the data for all the companies
ohlcv = {}
for ticker in tickers:
    try:
        # Pull the stock details for that ticker
        data[ticker] = api.get_aggs(ticker, 1, 'day', '2020-01-01', '2020-06-01').df
        # Calculate their daily return
        data[ticker]['daily return'] = data[ticker]['close'].pct_change()
        print('Pulling ohlcv data for {:s}'.format(ticker))
    except:
        # If nothing is found, throw error and continue
        print('Error encountered pulling ohlcv data for {:s}'.format(ticker))
```

```python
# See some summary statistics
print(data.info())
print(' ')
print(data.describe().round(2))
```

```python
# Can specify what statistics you want to see with the .aggregate() method
data.aggregate([min,
              np.mean,
              np.std,
              np.median,
              min])
```

```python
# Typically want to assess changes over time rather than the absolute values themselves - hence transformations are required
print(data.diff().head())  # Absolute differences
print(' ')
print(data.diff().mean())
```

```python
# Further, in financial analysis, the percentage changes are more desirable than the absolute changes as it's less sensitive to scale
print(data.pct_change().round(3).head())
data.pct_change().mean().plot(kind='bar')
```

```python
# Calculate log returns as follows:
rets = np.log( data / data.shift(1) )
print(rets.head().round(3))

rets.cumsum().apply(np.exp).plot(figsize=(12,10))
```

```python
# Resampling is common to move from tick level data to day close or something on a longer frequency
print(data.resample('1w', label='right').last().head())  # This is end of day data (That's what .last() does?)
print(' ')
print(data.resample('1m', label='right').last().head())  # End of day data on monthly intervals
print(' ')
rets.cumsum().apply(np.exp).resample('1m', label='right').last().plot(figsize=(10,8)) # Plots the resampled (month frequency) cumulative sum with exp applied
```

Important when resampling to specify label='right' to ensure that the right label of the interval is taken, as well as specifying the .last() entry of the interval, both to ensure you don't introduce a foresight bias

```python
# Rolling statistics.

# Consider only a single series of info
symbol = 'AAPL.O'
data = pd.DataFrame(data[symbol]).dropna()
print(data.tail())
print(' ')

# Define window over which the rolling stats will be applied
window = 20
data['min'] = data[symbol].rolling(window=window).min()
data['mean'] = data[symbol].rolling(window=window).mean()
data['std'] = data[symbol].rolling(window=window).std()
data['median'] = data[symbol].rolling(window=window).median()
data['max'] = data[symbol].rolling(window=window).max()
data['ewma'] = data[symbol].ewm(halflife=0.5, min_periods=window).mean()  # Exponentially weighted moving average

print(data.dropna().head())
```

```python
ax = data[['min', 'mean', 'max']].iloc[-200:].plot( figsize=(10, 6), style=['g--', 'r--', 'g--'], lw=0.8)
data[symbol].iloc[-200:].plot(ax=ax, lw=2.0)
#[-200:] means the last 200 values
```

```python
# Basic strategy involving rolling statistics says the investor goes long when the shorter average is larger than the longer and vice versa

# Create two averages
data['SMA1'] = data[symbol].rolling(window=42).mean()
data['SMA2'] = data[symbol].rolling(window=252).mean()
print(data[[symbol, 'SMA1', 'SMA2']].tail().round(2))
data[[symbol, 'SMA1', 'SMA2']].plot(figsize=(10,8))
plt.show()
```

```python
data.dropna(inplace=True)
data['Positions'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)  # where(condition, fillTrue, fillFalse)
ax = data[[symbol, 'SMA1', 'SMA2', 'Positions']].plot(figsize=(10, 8), secondary_y='Positions')  # This is how to get another y axis on the plot
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
```
