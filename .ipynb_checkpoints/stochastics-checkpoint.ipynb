## Simulation - random variables

Black scholes merton stock price setup:<br>
<br>
$S_T=S_0\mathrm{exp}\left[(r-\dfrac{\sigma^2}{2})T + \sigma Z\sqrt{T}\right]$
<br>
Where<br>
$S_T$ is the stock price at time T<br>
$T$ is the time horizon normally in years, or fractions of years
$S_0$ is the initial stock price<br>
$r$ is the risk free rate<br>
$\sigma$ is the volatility (standard deviation of the returns) of S<br>
$Z$ is a standard normal distribution

Note that the stochastic part of the above equation is on the time side of the brownian scaling with root time

```python
S0 = 100
r = 0.05
sigma = 0.25
T = 2
I = 10000  # Number of Monte carlo paths to simulate

ST1 = S0 * np.exp((r-(sigma**2)/2)*T + sigma*math.sqrt(T)*npr.standard_normal(I))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(ST1, bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
```

Above plot looks like a log normal distribution, so we can try using that in the simulation instead

```python
ST2 = S0 * npr.lognormal((r - 0.5*sigma**2)*T, sigma*math.sqrt(T), size=I)
plt.figure(figsize=(10, 6))
plt.hist(ST2, bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
```

Looks pretty much the same. Statistical similarity is assured by making a little comparitive function from scipy.stats

```python
from scipy import stats as scs
def print_statistics(a1, a2): 
    ''' Prints selected statistics.
    Parameters ==========
    a1, a2: ndarray objects 
    results objects from simulation
    ''' 
    sta1 = scs.describe(a1)
    sta2 = scs.describe(a2) 
    print('%14s %14s %14s' % ('statistic', 'data set 1', 'data set 2'))
    print(45 * "-") 
    print('%14s %14.3f %14.3f' % ('size', sta1[0], sta2[0])) 
    print('%14s %14.3f %14.3f' % ('min', sta1[1][0], sta2[1][0]))
    print('%14s %14.3f %14.3f' % ('max', sta1[1][1], sta2[1][1])) 
    print('%14s %14.3f %14.3f' % ('mean', sta1[2], sta2[2])) 
    print('%14s %14.3f %14.3f' % ('std', np.sqrt(sta1[3]), np.sqrt(sta2[3])))
    print('%14s %14.3f %14.3f' % ('skew', sta1[4], sta2[4])) 
    print('%14s %14.3f %14.3f' % ('kurtosis', sta1[5], sta2[5]))

print_statistics(ST1, ST2)
```

Differences in above statistics mainly caused by sampling error

## Geometric brownian motion

Can think of a stochastic process as a sequence of random variables. Generally, stochastic processes used in finance exhibit the markov property (memoryless) that says their value tomorrow only depends on their value today (and not their history).

Consider geometric brownian motion:
$\dfrac{dS_t}{S_t}= rdt + \sigma dZ_t$<br>
where $dZ_t$ is the standard brownian motion<br>
The values of $S_t$ are log-normally distributed whilst the values of $\dfrac{dS_t}{S_t}$ are normally distributed<br>
<br>
Geometric brownian motion can discretised exactly by a Euler scheme with $\Delta t$ as the fixed discretisation interval and $Z_t$ being the standard normal distribution:<br>
$S_t = S_0\mathrm{exp}\left[ (r - \dfrac{\sigma^2}{2})\Delta t + \sigma Z_t \sqrt{T} \right]$

```python
I = 10000  # Number of simulation paths
M = 50  # Number of time intervals
T = 2
dt = T / M
S = np.zeros((M+1, I))
S[0] = 100

for t in range(1, M+1):
    S[t] = S[t-1]*np.exp((r - 0.5*sigma**2)*dt + sigma*math.sqrt(dt)*npr.standard_normal(I))

plt.figure(figsize=(10, 6))
plt.hist(S[-1], bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
```

Following is a comparison of the statistics resulting from the dynamic simulation as well as from the static simulation. The fully vectorised version using the lognormal distribution was the static simulation.

The dynamic approach allows you to visualise the path taken to get to the final state (see figure below), whereas the static one only gives you the final output state. Sometimes you need the dynamic version to assess the behaviour of american or bermudan options that can be exercised at any given date, not just the maturity date

```python
print_statistics(S[-1], ST2)

plt.figure(figsize=(10, 6)) 
plt.plot(S[:, :10], lw=1.5) 
plt.xlabel('time') 
plt.ylabel('index level')
```

## Square root diffusion (a mean reversion process)

Stochastic differential equation for square root diffusion:<br>
$dx_t=\kappa(\theta-x_t)dt+\sigma dZ_t\sqrt{x_t}$<br>
where<br>
$x_t$ is the process level at date $t$<br>
$\kappa$ is the mean reversion factor<br>
$\theta$ is the long term mean of the process<br>
$\sigma$ is the volatility<br>
$Z_t$ is standard brownian motion<br>
<br>
The values of $x_t\sim \chi^2$.<br>
(Euler) discretisation allows a lot of financial models to be approximated by the normal distribution. This is exact for geometric brownian motion.<br>
For $x_t$ here, the Euler scheme is biased.<br>
<br>
Using $s=t-\Delta t$ and $x^+=\mathrm{max}(x,0)$ the Euler scheme applied to square root diffusion is:<br>
$\tilde{x_t}=\tilde{x_s}+\kappa(\theta -\tilde{x_s}^+)\Delta t + \sigma Z_t\sqrt{\tilde{x_s}^+}\sqrt{\Delta t}$<br>
$x_t = \tilde{x_t}^+$

```python
x0 = 0.05  # Initial value for e.g. a short rate
kappa = 3
theta = 0.02
sigma = 0.1
T = 2
I = 10000
M = 50
dt = T / M

def srd_euler():
    xh = np.zeros((M + 1, I)) 
    x = np.zeros_like(xh) 
    xh[0] = x0 
    x[0] = x0 
    for t in range(1, M + 1):
        xh[t] = (xh[t - 1] + kappa * (theta - np.maximum(xh[t - 1], 0)) * dt +
                 sigma * np.sqrt(np.maximum(xh[t - 1], 0)) * math.sqrt(dt) * npr.standard_normal(I))
    x = np.maximum(xh, 0) 
    return x

x1 = srd_euler()

plt.figure(figsize=(10, 6)) 
plt.hist(x1[-1], bins=50) 
plt.xlabel('value') 
plt.ylabel('frequency')
plt.title('Dynamically simulated square root diffusion at maturity (Euler scheme)')
plt.show()
```

Show the first 10 simulated paths

```python
plt.figure(figsize=(10, 6)) 
plt.plot(x1[:, :10], lw=1.5) 
plt.xlabel('time') 
plt.ylabel('index level')

# There is negative anual drift because x0 > theta
```

```python
# Because the Euler discretisation is exact only for geometric brownian motion, the above simulation has a statistical bias.
# This can be corrected for by using a version of the chi squared distribution that accounts for noncentrality:
# npr.noncentral_chisqure()
# That's done in the book page 557, but i can't be bothered to copy it into here as it looks basically the same.

# Note though that the exact scheme is computationally more intensive and takes approximately twice as long to execute
```

## Stochastic volatility

BSM major assumption is that of constant volatility, however in reality, the volatility itself is a stochastic process.

This has been accounted for in the model of Heston (1993) where the volatility is its own stochastic process characterised by brownian motion $dZ^2_t$ (stock price itself follows another stochastic process with brownian motion $dZ^1_t$)

$dS_t=rS_tdt+\sqrt{v_t}S_tdZ_t^1$<br>
$dv_t=\kappa_v(\theta_v - v_t)dt+\sigma_v\sqrt{v_t}dZ_t^2$<br>
$dZ_t^1\,dZ_t^2 = \rho$ as correlation.

Part of the model is calcualing the correlation between $dZ^1_t$ and $dZ^2_t$ which requires the cholskey decomposition of the correlation matrix. This allows you to account for the leverage effect, which basically says that volatility goes up in times of market stress (declining market), and down in a bull market

```python
S0 = 100
r = 0.05
v0 = 0.1  # Initial instantanious volatility value
kappa = 3
theta = 0.25
sigma = 0.1
rho = 0.6  # Fix the correlation between the two brownian motions
T = 1

corr_mat = np.zeros((2,2))
corr_mat[0, :] = [1.0, rho] 
corr_mat[1, :] = [rho, 1.0] 
cho_mat = np.linalg.cholesky(corr_mat)  # Choleskey decomp

cho_mat
```

```python
# Simulation of stochastic volatility
# Initialise by pre-generating all the random numbers required
v0 = 0.1
S0 = 100
M = 50
I = 10000
dt = T / M
ran_num = npr.standard_normal((2, M+1, I))  # two sets: The first one for the stochastic index (underlying price) process and the second one for the stochastic volatility process

v = np.zeros_like(ran_num[0])
vh = np.zeros_like(v)  # v history (Array that records the historical path. v will be overwritten with the max of this)
v[0] = v0
vh[0] = v0

# For volatility
for t in range(1, M+1):
    ran = np.dot(cho_mat, ran_num[:, t, :])   # Picks out the relevant random number subset for volatility and transforms it via cholsky
    
    vh[t] = (vh[t - 1] + kappa * (theta - np.maximum(vh[t - 1], 0)) * dt +  # Simulate paths based on (exact) euler scheme for geometric brownian motion
             sigma * np.sqrt(np.maximum(vh[t - 1], 0)) * math.sqrt(dt) * ran[1])

v = np.max(vh,0)

S = np.zeros_like(ran_num[0])
S[0] = S0 
for t in range(1, M + 1):
    ran = np.dot(cho_mat, ran_num[:, t, :])  # Picks out the random numbers for the index instead of the volatility
    S[t] = S[t - 1] * np.exp((r - 0.5 * v[t]) * dt + np.sqrt(v[t]) * ran[0] * np.sqrt(dt))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))
ax1.hist(S[-1], bins=50) 
ax1.set_xlabel('index level')
ax1.set_ylabel('frequency')
ax2.hist(vh[-1], bins=50) 
ax2.set_xlabel('volatility')
```

This illustrates another advantage of working with the Euler scheme for the square-root diffusion: correlation is easily and consistently accounted for since one only draws standard normally distributed random numbers. There is no simple way of achieving the same with a mixex approach (i.e., using Euler for the index and the noncentral chi-square-based exact approach for the volatility process)

```python
# Here we see that the volatility is drifting positively and that it, as expected, converges upon 0.25 (because this is
# a mean reverting process that we initialised with a long term mean of 0.25)

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6))
ax1.plot(S[:, :10], lw=1.5)
ax1.set_ylabel('index level') 
ax2.plot(vh[:, :10], lw=1.5) 
ax2.set_xlabel('time') 
ax2.set_ylabel('volatility')
```

So to summarise the above, we have taken the index level as geometric brownian motion (which can be exactly represented by euler discretisation), where the volatility in each simulated path was taken to be another stochastic process which was mean reverting (which was represented by a square root diffusion (mean reverting) process)

The result is an index level that can now reach a much larger range (higher max and lower min) of values at matrurity than if the volatility was held constant.

## Jump diffusion

Asset prices and volatility can jump around. in 1976 Merton published the jump diffusion model that allows for log-normally distributed jumps:

$dS_t=(r-r_j)S_tdt+\sigma S_tdZ_t+J_tS_tdN_t$<br>
where<br>
$S_t$ is the index level at time $t$<br>
$r$ is the constant riskless short rate<br>
$r_j=\lambda\left(\mathrm{exp}\left[ \mu_J+\dfrac{\delta^2}{2}\right]-1\right)$ is a drift correction for jumps to maintain risk neutralilty<br>
$\sigma$ is the constant volatility of $S$<br>
$Z_t$ is standard brownian motion<br>
$J_t$ is a jump at date $t$ with distribution $\mathrm{log}(1+J_t)\approx \Phi(\mathrm{log}(1+\mu_J)-\dfrac{\delta2}{2}, \delta^2)$<br>
$\quad$where $\Phi()$ is the CDF of the standard normal distribution<br>
$N_t\sim\mathrm{Poi}(\lambda)$<br>
<br>
<br>
The Euler discretisation of this is:<br>
$S_t=S_{t-\Delta t}\left( \mathrm{exp}\left[ (r-r_J-\dfrac{\sigma^2}{2})\Delta t+\sigma\sqrt{\Delta t}Z_t^1 \right] +\left( \mathrm{exp} \left[\mu_J + \delta Z_t^2\right]-1\right)y_t \right)$<br>
where<br>
$Z_t^i$ are standard normal distribution terms<br>
$y_t\sim\mathrm{Poi}(\lambda)$

```python
S0 = 100
r = 0.05
sigma = 0.2
lamb = 0.75  # Jump intensity
mu = -0.6  # Mean jump size
delta = 0.25  # Jump volatility
rj = lamb * (np.exp(mu+(delta**2)/2)-1)  # Drift correction

T = 1
M = 50
I = 10000
dt = T / M
```

```python
S = np.zeros((M + 1, I)) 
S[0] = S0 
sn1 = npr.standard_normal((M + 1, I)) 
sn2 = npr.standard_normal((M + 1, I)) 
poi = npr.poisson(lamb * dt, (M + 1, I)) 

for t in range(1, M + 1, 1):
    # Simulation based on the exact euler scheme
    S[t] = S[t - 1] * (np.exp((r - rj - 0.5 * sigma ** 2) * dt +
                              sigma * math.sqrt(dt) * sn1[t]) + (np.exp(mu + delta * sn2[t]) - 1) * poi[t])
    S[t] = np.maximum(S[t], 0)

plt.figure(figsize=(10, 6))
plt.hist(S[-1], bins=50)
plt.xlabel('value') 
plt.ylabel('frequency')
```

Above distribution has a second peak which is due to the jumps down in value. We can see these in the first 10 paths:

```python
plt.figure(figsize=(10, 6)) 
plt.plot(S[:, :10], lw=1.5) 
plt.xlabel('time') 
plt.ylabel('index level')
```

## Variance reduction

The random number generation in python can only make psudorandom numbers. These don't full represent the statistics we'd like. We can get closer to the desired 'perfect' distribution by only drawing half the random numbers, and then defining the remaining half as the negative of the first half. This is called the use of $\textbf{antithetic variates}$.

This means that by construction the random set has mean exactly equal to 0

```python
sn = npr.standard_normal(int(10000 / 2)) 
sn = np.concatenate((sn, -sn))

sn.mean()  # Machine precision or thereabouts
```

```python
print('%15s %15s' % ('Mean', 'Std. Deviation')) 
print(31 * "-") 
for i in range(1, 31, 2):
    npr.seed(1000) 
    sn = npr.standard_normal(i ** 2 * int(10000 / 2))
    sn = np.concatenate((sn, -sn))
    print("%15.12f %15.12f" % (sn.mean(), sn.std()))
```

Using these antithetic variates can fix the nonzero first moment of the psudorandom numbers, but it does't impact on the second moment. It's still not quite = 1.

Can do this by standardising the distribution

```python
sn = npr.standard_normal(10000)
sn_new = (sn - sn.mean()) / sn.std()

print(sn_new.mean())
print(sn_new.std())

# Now the mean is 0 (to within machine precision) and the standard deviation is 1 to within machine precision.
```

```python
def gen_sn(M, I, anti_paths=True, mo_match=True):
    ''' F?ncti?n t? generate rand?m n?mbers for sim?lati?n.
    Parameters
    ========== 
    M: int
        n?mber ?? time inter?als ??r discretizati?n 
    I: int
        n?mber ?? paths t? be sim?lated 
    anti_paths: b??lean
        use of antithetic variates 
    mo_math: b??lean
        use of moment matching
    '''
    if anti_paths is True:
        sn = npr.standard_normal((M + 1, int(I / 2)))
        sn = np.concatenate((sn, -sn), axis=1)
    else: sn = npr.standard_normal((M + 1, I))
    
    if mo_match is True:
        sn = (sn - sn.mean()) / sn.std()
    return sn
```

## Monte carlo evaluation of contingent claims (options, derivatives, hybrid instruments etc)

"Simply stated, in a risk-neutral world, the value of a contingent claim is the discounted expected payoff under the risk-neutral (martingale) measure" - Yves Hilpisch.

This means that the option price now = expected price at maturity discounted back to today.

### European Options:
Payoff only at maturity date according to:<br>
Put: $h(S_T)=\mathrm{max}(S_T-K,0)$ for strike price $K$<br>
Call: $h(S_T)=\mathrm{max}(K - S_T,0)$<br>
<br>
Price of a call option:<br>
$C_0=\mathrm{e}^{-rT}E_0^Q(h(S_T))=\mathrm{e}^{-rT}\int_0^{\infty}h(s)q(s)ds$ (Expectation value discounted back to today)<br>
This is solved via monte carlo simulation by running a large number of random paths to maturity, taking their average and discounting that value back to today.<br>
<br>
The corresponding monte carlo estimator is:<br>
$\tilde{C_0}=\mathrm{e}^{-rT}\dfrac{1}{I}\Sigma_{i=1}^Ih(\tilde{S}_T^i)$<br>
for total number of simulated paths $I$

```python
# For a stock price that varies according to geometric brownian motion consider
S0 = 100
r = 0.05
sigma = 0.25
T = 1
I = 50000

# Static implementation (only see end result, not the paths taken to get there)
def gbm_mcs_stat(K):
    ''' Valuation of European call option in Black-Scholes-Merton by
    Monte Carlo simulation (of index level at maturity)
    
    Parameters 
    ========== 
    K: float
        (positive) strike price of the option

    Returns
    =======
    C0: float
        estimated present value of European call option
    '''
    
    sn = gen_sn(1, I)
    # Simulate index level at maturity
    ST = S0 * np.exp((r - 0.5*sigma**2)*T + sigma*math.sqrt(T)*sn[1])
    # Calculate payoff at maturity
    hT = np.maximum(ST - K, 0)
    # Calculate monte carlo estimator
    C0 = math.exp(-r*T)*(np.mean(hT))
    return C0
```

```python
gbm_mcs_stat(K=105)
```

```python
# Dynamic version of the above
M = 50  # Number of time steps

def gbm_mcs_dyna(K, option='call'):
    ''' Valuation of European Options in Black-Scholes-Merton
    by Monte Carlo simulation (of index level paths)
    
    Parameters
    ==========
    K: float
        (positive) strike price of the option 
    option : string
        type of the option to be valued ('call', 'put')

    Returns
    =======
    C0: float
        estimated present value of European call option
    '''
    
    dt = T / M
    # Simulation of index paths
    S = np.zeros((M+1, I))
    S[0] = S0
    sn = gen_sn(M, I)
    for t in range(1, M+1):
        S[t] = S[t-1]*np.exp((r - 0.5*sigma**2)*dt + sigma*math.sqrt(dt)*sn[t])
    # Case based valuation of payoff
    if option == 'call':
        hT = np.maximum(S[-1]-K,0)
    if option == 'put':
        hT = np.maximum(K-S[-1],0)
    # Monte carlo estimator
    C0 = math.exp(-r*T)*np.mean(hT)
    return C0
```

```python
print(gbm_mcs_dyna(K=110., option='call'))
print(gbm_mcs_dyna(K=110., option='put'))
```

We can look at how closely these aggree to the analytical results by using the bsm_functions module

```python
from bsm_functions import bsm_call_value

stat_res = [] 
dyna_res = [] 
anal_res = []  # Haaaaaaaaaaaaaaaa
k_list = np.arange(80., 120.1, 5.)  # Range of strike prices
np.random.seed(100)

for K in k_list: 
    stat_res.append(gbm_mcs_stat(K))
    dyna_res.append(gbm_mcs_dyna(K))
    anal_res.append(bsm_call_value(S0, K, T, r, sigma))  # Haaaaaaa. This just implements the C0 = S0*Phi(d1) - K*exp(-2)Phi(d2)

stat_res = np.array(stat_res) 
dyna_res = np.array(dyna_res)
anal_res = np.array(anal_res)  # Haaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa
```

### American options (Can be exercised at any time between purchase and maturity)

This requires an optimal stopping problem to be solved instead of the simple simulation and discounting for European options

The optimal stopping problem for an American option is:<br>
$V_0=\textrm{supp}_{\tau\in(0,\Delta t,2\Delta t,...,T)}\mathrm{e}^{-rT}E_0^Q(h_\tau(S_\tau))$<br>
The value of the American option at time $t$ is:<br>
$V_t(s)=\mathrm{max}(h_t(s),C_t(s))$<br>
where $C_t(s)=E_t^Q(\mathrm{e}^{-r\Delta t}V_{t+\Delta t}(S_{t+\Delta t}) | S_t=s)$ is the continuation value of the option given an index value of $S_t=s$

### American options values calculated with least squares monte carlo (LSMC)

So to get a value for an American option, you need the continuation value.

So the problem is now: how to get the continuation value.

We define $Y_{t,i}=\mathrm{e}^{-r\Delta t}V_{t+\Delta t,i}$ to be the simulated continuation value for path $i$ at time $t$, where $I$ paths have been discretised into $M$ steps of $\Delta t$.

Can't use that number directly, because you'd be using info from the future, but you can use the values of each simulated path up to current.

The continuation value is then given by the regression estimate:<br>
$\hat{C}_{t,i}=\Sigma_{d=1}^D\alpha^{*}_{d,t}b_d(S_{t,i})$<br>
where the sum is over $D$ basis functions.

The optimal regression parameters $\alpha^{*}_{d,t}$ are the solutions to the least squares problem:<br>
$\mathrm{min}_{\alpha_{1,t},...,\alpha_{2,t}}\dfrac{1}{I}\Sigma_{i=1}^I\left(Y_{t,i}-\Sigma_{d=1}^D\alpha_{d,t}b_d(S_{t,i})\right)^2$

```python
# Code to implement least squares monte carlo for american options

def gbm_mcs_amer(K, option='call'):
    ''' Val?ati?n ?? American ?pti?n in Black-Sch?les-Mert?n by M?nte Carl? sim?lati?n by LSM alg?rithm

    Parameters 
    ========== 
    K : ?l?at
        (p?siti?e) strike price ?? the ?pti?n
    ?pti?n : string
        type ?? the ?pti?n t? be ?al?ed ('call', 'p?t')
    
    Ret?rns 
    ======= 
    C0 : ?l?at
        estimated present ?al?e ?? American call ?pti?n
    '''
    dt = T / M
    df = math.exp(-r * dt)
    # simulation of index levels
    S = np.zeros((M + 1, I))
    S[0] = S0
    sn = gen_sn(M, I)
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt
                + sigma * math.sqrt(dt) * sn[t])
    # case based calculation of payoff
    if option == 'call':
        h = np.maximum(S - K, 0)
    else:
        h = np.maximum(K - S, 0)
    # LSM algorithm
    V = np.copy(h)
    for t in range(M - 1, 0, -1):
        reg = np.polyfit(S[t], V[t + 1] * df, 7)
        C = np.polyval(reg, S[t])
        V[t] = np.where(C > h[t], V[t + 1] * df, h[t])  # np.where(condition, filltrue, fillfalse)
    # MCS estimator
    C0 = df * np.mean(V[1])
    return C0
```

```python
print(gbm_mcs_amer(110., option='call'))
print(gbm_mcs_amer(110., option='put'))
```

Given that the price of an option decreases as the time to maturity decreases, the European option value represents a lower bound on the American option value.

The increased value of the American option over the European one prior to maturity is known as the $\textbf{early exercise premium}$

You can estimate the value of the early exercise premium by comparing european to american options. Here is an example looking at puts:

```python
euro_res = [] 
amer_res = []
k_list = np.arange(80., 120.1, 5.)
for K in k_list:
    euro_res.append(gbm_mcs_dyna(K, 'put'))
    amer_res.append(gbm_mcs_amer(K, 'put'))

euro_res = np.array(euro_res) 
amer_res = np.array(amer_res)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6))
ax1.plot(k_list, euro_res, 'b', label='European put')
ax1.plot(k_list, amer_res, 'ro', label='American put') 
ax1.set_ylabel('call option value')
ax1.legend(loc=0)
wi = 1.0 
ax2.bar(k_list - wi / 2, (amer_res - euro_res) / euro_res * 100, wi) 
ax2.set_xlabel('strike') 
ax2.set_ylabel('early exercise premium in %') 
ax2.set_xlim(left=75, right=125);

# So the early exercise premium can be up to approximately 10% in this example.
```
